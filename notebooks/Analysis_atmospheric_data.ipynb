```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.figure(figsize=(8, 5))
plt.scatter(
    exterior_df["Air Temp"],
    exterior_df["Specific Humidity"],
    color="blue", edgecolor="k", alpha=0.7, s=60
)

plt.title("Specific Humidity vs Air Temperature", fontsize=14)
plt.xlabel("Air Temperature (C)", fontsize=12)
plt.ylabel("Specific Humidity (g/kg)", fontsize=12)

# Limit number of ticks
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(8))
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(8))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df = exterior_df
df = df[df["Report"] != "David_Bly_NW6"]

def specific_and_relative_summary(df):

    import matplotlib.pyplot as plt

    # Create a figure and primary y-axis
    fig, ax1 = plt.subplots(figsize=(8, 5))

    # Plot Specific Humidity on the primary y-axis (left)
    ax1.set_xlabel("Air Temperature (C)")
    ax1.set_ylabel("Specific Humidity (g/kg)", color="blue")
    ax1.scatter(df["Air Temp"], df["Specific Humidity"], color="blue", marker='o', label="Specific Humidity")
    ax1.tick_params(axis='y', labelcolor="blue")

    # Create a secondary y-axis (right) sharing the same x-axis
    ax2 = ax1.twinx()
    ax2.set_ylabel("Relative Humidity (%)", color="red")
    ax2.scatter(df["Air Temp"], df["Relative Humidity"], color="red", marker='s', label="Relative Humidity")
    ax2.tick_params(axis='y', labelcolor="red")

    # Add a title and grid
    plt.title("Specific vs Relative Humidity by Air Temperature")
    fig.tight_layout()
    plt.grid(True)
    plt.show()

specific_and_relative_summary(df)
```

```python
#The property with the highest relative humidity indoors:
LU1 = atmospheric_data_df[atmospheric_data_df['Report']=='David_Bly_LU1']
E7 = atmospheric_data_df[atmospheric_data_df['Report']=='David_Bly_E7']
SO50 = atmospheric_data_df[atmospheric_data_df['Report']=='David_Bly_SO50']
```

```python
def summary_plot_individual_property(df):
    fig, ax1 = plt.subplots(figsize=(12,8))

    # Primary axis: Relative Humidity
    ax1.scatter(df["Air Temp"], df["Relative Humidity"],
                s=100, c="skyblue", edgecolor="k", label="Relative Humidity (%)")

    for i, loc in enumerate(df["Location"]):
        ax1.text(df["Air Temp"].iloc[i] + 0.1, df["Relative Humidity"].iloc[i], loc, fontsize=9)

    ax1.set_xlabel("Air Temperature (C)")
    ax1.set_ylabel("Relative Humidity (%)", color="blue")
    ax1.tick_params(axis="y", labelcolor="blue")

    # Secondary axis: Specific Humidity
    ax2 = ax1.twinx()
    ax2.scatter(df["Air Temp"], df["Specific Humidity"],
                s=100, c="orange", edgecolor="k", label="Specific Humidity (g/kg)")
    ax2.set_ylabel("Specific Humidity (g/kg)", color="orange")
    ax2.tick_params(axis="y", labelcolor="orange")

    # Title & Grid
    fig.suptitle("Humidity vs Temperature by Location", fontsize=14)
    ax1.grid(True, linestyle="--", alpha=0.6)

    # Combine legends and move them outside
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2,
            loc="center left", bbox_to_anchor=(1.05, 0.5))

    # Adjust layout to fit the legend
    plt.tight_layout()
    plt.show()
```

```python
summary_plot_individual_property(LU1)
```

```python
summary_plot_individual_property(SO50)
```
