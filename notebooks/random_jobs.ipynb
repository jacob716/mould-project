```python
import pandas as pd

# Load both CSVs
df1 = pd.read_csv("../csv_files/merged_reports_categorised.csv")
df2 = pd.read_csv("../csv_files/merged_reports_categorised_batch2.csv")
mapping = pd.read_csv("../csv_files/mapping.csv")  # contains id, new_name, old_name

# Merge df1 with mapping to replace filenames
df1 = df1.merge(mapping, left_on="filename", right_on="old_name", how="left")

# Overwrite filename column with the new name
df1["filename"] = df1["new_name"]

# Drop helper columns (optional)
df1 = df1.drop(columns=["old_name", "new_name", "id"])

# Concatenate both
combined = pd.concat([df1, df2], ignore_index=True)

# Save combined file
combined.to_csv("../csv_files/merged_reports_categorised_all.csv", index=False)

print("✅ Combined CSV saved with updated filenames")
```
